# Built-in Functions ที่ควรรู้

## `abs(args)`
ส่งออกเป็นค่าสัมบูรณ์ของค่าที่รับเข้าไป

**Input** - `args` เป็น `float` หรือ `int`

**Output** - `float` หรือ `int` เป็นค่าสัมบูรณ์ของค่านั้นๆ

```python
abs(-12)
```

```python
abs(-112.44)
```

----
## `divmod(args1, args2)`
หาร `args1` ด้วย `args2` และส่งออกเป็นผลหารและเศษ

**Input**
* `args1` - เศษ (`float` หรือ `int`)
* `args2` - ส่วน (`float` หรือ `int`)

**Output** - `tuple` ในรูปแบบ `(ผลหาร, เศษ)`

```python
divmod(4,2)
```

```python
divmod(7,2)
```

## `len(args)`
ส่งออกค่าของความยาวค่าที่ใส่เข้าไป

**Input** - `args` (`string`, `tuple`, `list` หรือ `range`)

**Output** - `int` เป็นความยาวของ input

```python
len("tornado")
```

```python
jingles = ["died in a tornado", "the sun is a deadly lazer", "society"]
len(jingles)
```

```python
len(range(0,5))
```

## `max(args)`
หาค่าที่มากที่สุดใน `iterable` ที่ใส่เข้ามา

**Input** - `args` `iterable` (`tuple`, `list` หรือ `dict`)

**Output** - ค่าที่มากที่สุดใน Input

```python
year = [1917, 1991, 1939, 1945]
max(year)
```

```python
text = ["a", "b", "e", "d"] # e เป็นตัวอักษรตัวที่ 5 ซึ่งมากกว่าทุกตัว
max(text)
```

## `sorted(args, key=None, reverse=False)`
เรียงค่าใน `list` จากน้อยไปมาก

**Input**
* `args` - `iterable` (`tuple`, `list` หรือ `dict`)
* `key=`- `None` or (Ex: `str.lower`, `lambda i: len(i)`) (Optional)
* `reverse=` - Boolean `False` or `True` (Optional)

**Output** - `list`

```python
mylist = [4, 7, 1, 9, 14, 5, 7]
sorted(mylist)
```

```python
ourlist = ["slavsya", "otechestvo", "nashe", "svobodnoye"]
sorted(ourlist, key=len)
```

```python
mytuple = (1, 9, 10, 19, 17)
sorted(mytuple, reverse=True)
```
