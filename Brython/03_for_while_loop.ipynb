# `while`

`while` คือลูปชนิดหนึ่งที่จะทำงานซ้ำๆจนกว่าเงื่อนไขจะไม่เป็น `True`

**หมายเหตุ** - หากใช้เป็น `while True` ตัวลูปนั้นจะทำงานไม่รู้จบจนจนกว่าจะมีการใช้คำสั่ง `break`

ตัวอย่างการใช้งาน while loop ในการแสดงคำว่า "Yes" 5 ครั้ง

```python
i = 1
while i <= 5:
    print("Yes")
    i += 1
```

จากตัวอย่างตอนแรกกำหนดใน `i` มีค่าเป็น 1 และตัวลูปให้เงื่อนไขเป็น `i <= 5`
เมื่อเริ่มการทำงานจะเช็คเงื่อนไขว่าเป็นจริงหรือไม่หากเป็นจริงให้ทำงานต่อ
* 1 <= 5 หรือไม่? | ใช่ | ทำการรันโค้ดด้านใน (แสดงคำว่า "Yes" และเพิ่มค่า i ไปอีก 1 เป็น 2)
* 2 <= 5 หรือไม่? | ใช่ | ทำการรันโค้ดด้านใน (แสดงคำว่า "Yes" และเพิ่มค่า i ไปอีก 1 เป็น 3)
* 3 <= 5 หรือไม่? | ใช่ | ทำการรันโค้ดด้านใน (แสดงคำว่า "Yes" และเพิ่มค่า i ไปอีก 1 เป็น 4)
* 4 <= 5 หรือไม่? | ใช่ | ทำการรันโค้ดด้านใน (แสดงคำว่า "Yes" และเพิ่มค่า i ไปอีก 1 เป็น 5)
* 5 <= 5 หรือไม่? | ใช่ | ทำการรันโค้ดด้านใน (แสดงคำว่า "Yes" และเพิ่มค่า i ไปอีก 1 เป็น 6)
* 6 <= 5 หรือไม่? | ไม่ | จบลูปทันที และจะไม่มีการรันโค้ดด้านใน

จากด้านบนจะเขียนเป็นตารางได้เป็นดังนี้

| i | i <= 5 |
|---|--------|
| 1 | True   |
| 2 | True   |
| 3 | True   |
| 4 | True   |
| 5 | True   |
| 6 | False  |

สำหรับค่า `i` ที่เอาไว้ตรวจสอบเงื่อนไขก็สามารถเอามาใช้งานได้เช่นกันดังตัวอย่างนี้ที่จะเป็นการแสดงค่า `i` ออกมา

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

# `for`
`for` เป็นลูปอีกชนิดหนึ่งที่มีการกำหนดจำนวนครั้งการลูปที่แน่นอน โดยจะเป็นการลูปผ่านตัวอักษร(string), ไอเทมในลิสต์ หรือ ตัวเลขก็ได้

**ตัวอย่างการลูปผ่าน `string`**

```python
city = "Moscow"
for i in city:
    print(i)
```

ค่า `i` ในการลูปนั้นจะเป็นตัวอักษรตามตำแหน่งนั้นๆ โดยสั่งให้ทุกครั้งที่ลูปแสดงค่า `i` หรือตัวอักษรนั้นๆ
จากค่าที่ได้ใส่ไปคือ `city` ซึ่งมีค่าคือ "Moscow" เมื่อลูปรอบแรกค่า `i` จะเป็น "M"  ทำให้แสดงผลเป็น "M" ออกมา และ "o", "s", "c", "o", "w" ไปตามลำดับ

**ตัวอย่างการลูปผ่านลิสต์**

```python
cities = ["Kaliningrad", "St.Petersburg", "Volgograd"]
for i in cities:
    print(i)
```

เหมือนในตัวอย่างของการลูปผ่านตัวอักษร แต่จะเป็นการลูปผ่านไอเทมแต่ละตัวในลิสต์แทน

**ตัวอย่างการลูปผ่านตัวเลขโดยใช้ `range()`**

```python
for i in range(5):
    print(i)
```

การ `range()` จะเป็นการสั่งให้ลูปเป็นจำนวนกี่ครั้ง โดยการนับจะเริ่มที่ 0 เช่น `range(3)` จะเป็นค่า 0 - 2 ไม่ใช่ 1 - 3
โดยค่า `i` ที่ได้ก็จะนำมาใช้ได้เหมือนกัน

# การใช้ `break` เพื่อจบลูป
ลูปสามารถถูกสั่งให้จบได้โดยการใ้คำสั่ง `break` โดยเงื่อนไขไม่จำเป็นต้องเป็น False ดังตัวอย่าง

```python
while True:
    txt = input()
    if txt == "Stop":
        break
    print("[Output] You entered: " + txt)
```

จากตัวอย่างด้านบนเรากำหนดให้เงื่อนไขเป็น `True` ซึ่งหมายความว่าตัวลูปนั้นจะลูปไปเรื่อยๆไม่รู้จบและภายในลูปให้รับค่าและเก็บไว้ในตัวแปร `txt` จากนั้นมีเงื่อนไข `if` ตรวจหากว่าค่าที่ป้อนเข้ามาเป็น "Stop" ให้ทำการ `break` คำสั่งต่อไปคือให้แสดงผลข้อความ "[Output] You entered: " กับค่าของ `txt` ที่ได้ใส่เข้าไป

3 รอบแรกได้ใส่ตัวหนังสือปกติเข้าไปซึ่งไม่เข้าเงื่อนไข "Stop" จึงทำให้มีการแสดงค่าออกมาเรื่อยๆ จนกระทั่งบรรทัดสุดท้ายได้ใส่ค่าว่า "Stop" เข้าไปทำให้เข้าเงื่อนไขและทำการรันคำสั่ง `break` และลูปจะจบทันที โดยคำสั่งทุกอย่างที่อยู่ถัดไปก็จะไม่ทำงานเช่นกัน ทำให้ไม่มีการแสดงค่า "[Output] You entered: Stop" ออกมา

# การใช้ `continue`
`continue` เป็นการข้ามการทำงานของลูปแต่จะยังไม่เป็นการจบลูปทันที

```python
fruits = ["banana", "apple", "melon", "grape"]
for i in fruits:
    if i == "melon":
        continue
    print(i)
```

จากตัวอย่างด้านบนจะเป็นการใช้ for ลูปผ่านลิสต์ `fruits` และให้แสดงค่าออกมาทุกครั้งที่ลูป แต่ได้ใส่เงื่อนไขไว้ว่าหาก `i` เป็น "melon" ให้ทำการ `continue`

เมื่อเริ่มการทำงาจะมีการแสดงค่า "banana" และ "apple" ตามลำดับปกติเพราะไม่เข้าเงื่อนไข
แต่พอค่า `i` เป็น "melon" ตัวคำสั่ง `continue` จะงานและข้ามคำสั่งทั้งหมดที่อยุ่ถัดไปทั้งหมดซึ่งในที่นี้เป็นการแสดงค่า และลูปต่อ
หลังจากนั้น `i` เป็น "grape" ซึ่งไม่เข้าเงื่อนไขก็จะทำให้มีการแสดงค่าตามปกติ
