# Order of Operations - ลำดับการดำเนินการ
## PEMDAS
1. **P**arentheses - วงเล็บ | (x)
2. **E**xponent - เลขยกกำลัง | x<sup>2</sup>
3. **M**ultiplication and **D**ivision - การหาร และคูณ | 2 * 3, 12 / 8 _(ทำจากซ้ายไปขวา)_
4. **A**ddition and **S**ubtraction - การบวกและ ลบ | 1 + 9, 22 - 6 *(ทำจากซ้ายไปขวา)*

## Example

### 3 + 3 * 3 - 3 + 3
1. 3 + **3 * 3** - 3 + 3 (คูณและหาร)
2. 3 + 9 - 3 + 3 (บวกและลบ)
   1. **3 + 9** - 3 + 3 (หากมีหลายตัวทำจากซ้ายมาขวา)
   2. **12 - 3** + 3
   3. **9 + 3**

**Result: 12**


```python
3 + 3 * 3 - 3 + 3
```

### 30 - 12 / (6 - 2) + 6 * 2<sup>3</sup>
1. 30 - 12 / **(6 - 2)** + 6 x 2<sup>3</sup> (ในวงเล็บ)
2. 30 - 12 / 4 + 6 * **2<sup>3</sup>** (ตัวยกกำลัง)
3. 30 - 12 / 4 + 6 * 8 (คูณและหาร)
   1. 30 - **12 / 4** + 6 * 8 (หากมีหลายตัวทำจากซ้ายมาขวา)
   2. 30 - 3 + **6 * 8**
   3. 30 - 3 + 48
4. 30 - 3 + 48 (บวกและลบ)
   1. **30 - 3** + 48 (หากมีหลายตัวทำจากซ้ายมาขวา)
   2. **27 + 48**

**Result: 75**

```python
30 - 12 / (6 - 2) + 6 * 2 ** 3
```

# Variable

Varible หรือตัวแปรใน Python จะมีหลักๆอยู่ 3 ประเภทดังนี้
* `int` - จำนวนเต็ม เช่น `44` `112` `1945`
* `float` - จำนวนจริง เช่น `3.14` `291.02`
* `string` - ตัวหนังสือ เช่น `Hello There!` `General Kenobi!`

โดยสามารถใช้คำสั่ง `type()` ในการตรวจสอบประเภทตัวแปรได้

```python
print(type(123))
print(type(356.23))
print(type("Senate"))
```

ตัวแปรสามารถกำหนดชื่อให้มันได้ดังตัวอย่างนี้

```python
num = 5
lat = 1.024561
name = "Ben"
print(num)
print(lat)
print(name)
```

การเรียกใช้งานตัวแปรนั้นไม่สามารถเรียกใช้งานก่อนมีการประกาศตัวแปรได้ดังตัวอย่างนี้

```python
a = 1
print(a)
print(b)
b = 2
```

ดังตัวอย่างเมื่อมีการเรียกค่า `b` จะได้รับแจ้งว่า `NameError` เพราะยังไม่ได้ประกาศและกำหนดค่าให้ตัวแปร `b` ทำให้ไม่สามารถเรียกใช้ได้
